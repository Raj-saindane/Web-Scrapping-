```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.britannica.com/topic/Presidents-of-the-United-States-1846696"
```

```python
r=requests.get(url)
```

```python
r
```

```python
soup=BeautifulSoup(r.content,"html.parser")
```

```python
soup
```

```python
table=soup.find("table")
table
```

```python
title=soup.find("title")
title.text
```

```python
rows=table.find_all("tr")
rows
```

```python
for i in table.find_all("tr"):
    print(i)
```

```python
data=[]
for i,item in enumerate(rows):
    data.append(item.text.strip().split("\n"))
```

```python
data=data[3:]
print(data)
```

```python
import re
```

```python
#pip install pandas
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data=data)
df
```

```python
df.drop(df.columns[[1,2,4,5,7,8,10,11]],axis=1,inplace=True)
df
```

```python
df.rename(columns={0:"number",3:"president",6:"birthplace",9:"political_party",12:"term"},inplace=True)
df
```

```python
df.to_csv("List_of_US_Presidents.csv",index=False)
```
